```python
import yfinance as yf

# Create a Ticker object for NVIDIA
nvidia = yf.Ticker("NVDA")

# Get basic info about NVIDIA
info = nvidia.info
print("NVIDIA Info:")
print(info)

# Get historical market data for the last 5 days
history = nvidia.history(period="5d")
print("\nHistorical Data (last 5 days):")
print(history)
```
